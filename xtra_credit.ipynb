```python
from datetime import datetime
from datetime import timedelta
```

```python
class Library:  #parent class
    total=7 #global variable
    def __init__(self, quantity, price,category, slot): 
        self.quantity = quantity
        self.__price = price
        self.category=category
        self.slot=slot
        self.__day=None
    
    #returns date and time for returning scooters
    def setReturnDatetime(self):
        if self.category==1:
            today = datetime.now()+(self.slot*timedelta(hours=1))
            self.__day=today.strftime("%m/%d/%Y, %H:%M:%S")
            return self.__day
        elif self.category==2:
            today = datetime.now()+(self.slot*timedelta(days=1))
            self.__day=today.strftime("%m/%d/%Y, %H:%M:%S")
            return self.__day
        elif self.category==3:
            today = datetime.now()+(self.slot*timedelta(days=7))
            self.__day=today.strftime("%m/%d/%Y, %H:%M:%S")
            return self.__day
        
    def get_price(self):     #Encapsulation   
        return self.__price * self.quantity * self.slot
    
    
    #returnes available number of scooters
    def AvailableCount(self): #Encapsulation
        if self.total>=self.quantity:
            self.total=self.total-self.quantity
        else:
            print(f"{self.total} books left.")
        return self.total
    
        
    def __repr__(self): 
        return f'''Quantity: {self.quantity}, Price: {self.get_price()} for {self.slot} slots. 
                    There are only {self.__total} books available.'''
    
    #class method to set global variable
    @classmethod
    def update_total(cls,qty,oper):
        if oper=="add":
            cls.total=cls.total+qty
        elif oper=="sub":
            cls.total=cls.total-qty
        
#when someone rents scooter
class BookRent(Library): #child class
    def __init__(self, quantity, price, category, slot, name): 
        super().__init__(quantity, price, category, slot)
        self.name = name
        
    def __repr__(self): #polymorphism
        return f'''
                   UTD Library Rental Shop:
        
                   Name: {self.name} 
                   Quantity: {self.quantity} 
                   Price: ${self.get_price()}                     Slots for each: {self.slot} slots.
                   Return time: {self.setReturnDatetime()}.    Current time: {datetime.now().strftime("%m/%d/%Y, %H:%M:%S")}
                   
                   Show this slip before returning!'''

#when someone returns scooter
class BookReturn(Library): #child class
    def __init__(self, quantity, price, category, slot): 
        super().__init__(quantity, price, category, slot)
                            
    def __repr__(self): #polymorphism
        return f'''
            UTD Library Rental Shop:
        
            Quantity: {self.quantity} for {self.slot} slots.
            
            Available books {self.total}
            Thanks for returning!
            Visit Again!!'''
    
def pricePoint(n):
    if n==1:
        return 1
    elif n==2:
        return 2
    elif n==3:
        return 5
    else:
        return 0

def main():
    while(True): #loop for continuous interaction
        msg=''' 
            please choose the number for your activity
                1) Rent a book
                2) Return a book
                3) Exit 
            '''
        print(msg)
        choice=int(input("Enter your choice: "))
        
        if choice==3:
            print("Thanks for visiting us.")
            break
        
        elif choice==1:
            nm=input("enter your name: ")
            msg2=''' 
            please choose the number for your Duration to rent:
                
                1) Hourly
                2) Daily
                3) Weekly 
                
                '''
            print(msg2)
            t=int(input("enter the option number for the rental duration: "))
            s=int(input("enter the number of hours/days/weeks: "))
            q=int(input("enter the quantity of books: "))
            r=pricePoint(t)
            d1=BookRent(q,r,t,s,nm)
            receipt=str(d1)
            now = datetime.now()
            current_time = now.strftime("%H_%M_%S")
            op="Receipt_"+nm+"_"+current_time+".txt"
            with open(op, 'w') as f:
                f.write(receipt)
            f.close()
            print("The reciept is in your folder!")
            print(f'Now available book count is {d1.AvailableCount()}')
            Library.update_total(q,"sub")
        
        elif choice==2:
            q=int(input("enter the quantity of books for return: "))
            msg2=''' 
            please choose the number for your Duration
                
                1) Hourly
                2) Daily
                3) Weekly 
                
                '''
            print(msg2)
            t=int(input("enter the option number for the rental duration: "))
            s=int(input("enter the number of hours/days/weeks: "))
            r=pricePoint(t)
            d2=BookReturn(q,r,t,s)
            Library.update_total(q,"add")
            print(d2)        
        else:
            print("Invalid choice")
```

```python
import time
# get the start time
st_elapsed_time = time.time()
st_cpu_time = time.process_time()
main()
# get the end time
et_elapsed_time = time.time()
et_cpu_time = time.process_time()
# get the execution time
elapsed_time = et_elapsed_time - st_elapsed_time
cpu_time = et_cpu_time - st_cpu_time
print(f'Elapsed Execution time: {elapsed_time} seconds')
print(f'CPU Execution time: {cpu_time} seconds')
```
